```python
import json
import pandas as pd

def get_reactions(reacts):
    if type(reacts) == list:
        return len(reacts)
    else:
        return 0

final = pd.DataFrame()
    
file_path = list('/Users/extre_000/Downloads/facebook-jadehsuih/your_activity_across_facebook/messages/inbox/zinglusqurlyats_1977445935689959/message_1.json')

for i in range(33):
    print("".join(file_path))
    repath = "".join(file_path)

    with open(repath) as file:
        chat_history = json.load(file)
    
    messages = pd.DataFrame(chat_history['messages'])
    messages['rnum'] = messages['reactions'].apply(get_reactions)

    info = messages[(messages['reactions'].apply(get_reactions) >= 10)][['sender_name','content','rnum']]
    final = pd.concat([final,info])
    print(final)

    file_path[132] = str(int(file_path[132]) + 1)
```
